```python
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.morphology import binary_erosion, rectangle
import matplotlib.pylab as pylab

# Hm h tr  hin th nh vi tiu 
def plot_image(image, title=''):
    pylab.title(title, size=20)
    pylab.imshow(image)
    pylab.axis('off')  # n cc trc ca biu  (c th b dng ny nu mun hin th trc)

# c nh v chuyn sang grayscale
image_path = 'D:\\Learn DL\\X l nh\\BTL\\Sandipan_Dey_2018_Sample_Images\\images\\clock2.jpg'
im = rgb2gray(imread(image_path))

# Chuyn i nh grayscale sang nh nh phn da trn ngng c nh 0.5
im[im <= 0.5] = 0  # Cc gi tr <= 0.5 chuyn thnh 0 (background)
im[im > 0.5] = 1   # Cc gi tr > 0.5 chuyn thnh 1 (foreground)

# Hin th nh gc v cc kt qu xi mn
pylab.gray()  # Hin th nh  ch  grayscale
pylab.figure(figsize=(20, 10))

# Hin th nh gc
pylab.subplot(1, 3, 1)
plot_image(im, 'Original Image')

# Thc hin php xi mn vi hnh ch nht (1, 5)
im_eroded_1 = binary_erosion(im, rectangle(1, 5))
pylab.subplot(1, 3, 2)
plot_image(im_eroded_1, 'Erosion with Rectangle (1, 5)')

# Thc hin php xi mn vi hnh ch nht (1, 15)
im_eroded_2 = binary_erosion(im, rectangle(1, 15))
pylab.subplot(1, 3, 3)
plot_image(im_eroded_2, 'Erosion with Rectangle (1, 15)')

# Hin th kt qu
pylab.show()
```

```python
from skimage.morphology import skeletonize  # Nhp hm skeletonize t th vin scikit-image
from skimage.io import imread
from skimage import img_as_float
import matplotlib.pylab as pylab

# Hm  hin th nh gc v nh skeleton cnh nhau
def plot_images_horizontally(original, filtered, filter_name, sz=(18,7)):
    pylab.gray()  # Thit lp ch  mu xm cho nh
    pylab.figure(figsize = sz)  # Thit lp kch thc nh
    pylab.subplot(1, 2, 1)  # To subplot u tin  hin th nh gc
    plot_image(original, 'Original')  # Hin th nh gc
    pylab.subplot(1, 2, 2)  # To subplot th hai  hin th nh skeleton
    plot_image(filtered, filter_name)  # Hin th nh  skeletonize
    pylab.show()  # Hin th nh

# c nh v chuyn i nh thnh dng nh phn
im = img_as_float(imread('D:\\Learn DL\\X l nh\\BTL\\Sandipan_Dey_2018_Sample_Images\\images\\dynasaur.png')[..., 3])  # c nh v ly knh alpha (4 knh)
threshold = 0.5  # t ngng  chuyn nh sang nh phn
im[im <= threshold] = 0  # Cc pixel nh hn hoc bng ngng c gn thnh 0
im[im > threshold] = 1  # Cc pixel ln hn ngng c gn thnh 1

# p dng thut ton skeletonize
skeleton = skeletonize(im)

# Hin th nh gc v nh skeleton cnh nhau
plot_images_horizontally(im, skeleton, 'Skeleton', sz=(18, 9))
```
